```python
import librosa #analyse audio files
import soundfile #reading and writing audio files
import tensorflow as tf #building and training neural networks
import os, glob, pickle #file system operations, finding file with pattern
# pickle- serialisation and deserialisation of python objrcts
import numpy as np # scientific computing
from sklearn.model_selection import train_test_split # splitting dataset
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from sklearn.model_selection import GridSearchCV #hyperparameter tuning
from sklearn.model_selection import ShuffleSplit #cross validation
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from keras.models import Sequential #building sequential network
from keras.layers import Dense, Dropout #define layers in neural networks
from sklearn.preprocessing import LabelEncoder, StandardScaler #encoding categorical data, scaling numerical daya
from sklearn.metrics import accuracy_score, precision_recall_fscore_support #evaluate performance of the models
```

```python
def extract_feature(file_name, mfcc, chroma, mel): #used to extract features from an audio file
    with soundfile.SoundFile(file_name) as sound_file: #opens specified audio file
        X = sound_file.read(dtype="float32") #assigned audio data
        sample_rate=sound_file.samplerate #assigned sample rate data
        if chroma:
            stft=np.abs(librosa.stft(X)) #short time fourier transformation
        result=np.array([]) #to store extracted features
        if mfcc:
            mfccs=np.mean(librosa.feature.mfcc(y=X, sr=sample_rate, n_mfcc=40).T, axis=0)
            result=np.hstack((result, mfccs))
        if chroma:
            chroma=np.mean(librosa.feature.chroma_stft(S=stft, sr=sample_rate).T,axis=0)
            result=np.hstack((result, chroma))
        if mel:
            mel=np.mean(librosa.feature.melspectrogram(y=X, sr=sample_rate, n_mels=128).T,axis=0)
            result=np.hstack((result, mel))
    return result
```

```python
emotions={
  '01':'neutral',
  '02':'calm',
  '03':'happy',
  '04':'sad',
  '05':'angry',
  '06':'fearful',
  '07':'disgust',
  '08':'surprised'
}

observed_emotions=['calm', 'happy', 'fearful', 'disgust']
```

```python
def load_data(test_size):
    x,y=[],[]
    for file in glob.glob("/content/ravdess/Actor_*/*.wav"):
        file_name=os.path.basename(file)
        #print(file_name)
        emotion=emotions[file_name.split("-")[2]]
        if emotion not in observed_emotions:
            continue
        feature=extract_feature(file, mfcc=True, chroma=True, mel=True)
        x.append(feature)
        y.append(emotion)
    return train_test_split(np.array(x), y, test_size=test_size, random_state=9)
```

```python
x_train,x_test,y_train,y_test=load_data(test_size=0.3)
```

```python
parameter_space = {
    'hidden_layer_sizes': [(20)],
    'activation': ['logistic'],
    'solver': ['adam'],
    'alpha': [0.9],
    'learning_rate': ['adaptive'],
    'max_iter':[500],
    'random_state':[13]
}

model1 = GridSearchCV(model, parameter_space, n_jobs=-1, cv=cv)
```

```python
x_mlp,y_mlp=[],[]
for file in glob.glob("/content/ravdess/Actor_*/*.wav"):
  file_name=os.path.basename(file)
  #print(file_name)
  emotion=emotions[file_name.split("-")[2]]
  if emotion not in observed_emotions:
    continue
  feature=extract_feature(file, mfcc=True, chroma=True, mel=True)
  x_mlp.append(feature)
  y_mlp.append(emotion)
```

```python
#data sorting in MLP classifier

encoder = LabelEncoder()
y_mlp = encoder.fit_transform(y_mlp) #encode categorical data into numerical variables
scaler = StandardScaler() #standardizes features by removing mean and scaling variance to unit variance
x_mlp = scaler.fit_transform(x_mlp) #standardizes input features, applies transofrmation to both test and train sets
x_train, x_test, y_train, y_test = train_test_split(np.array(x_mlp), np.array(y_mlp), test_size=0.2, random_state=42)
model = Sequential() #linear stack of Neural network layers
#defining model architecture by adding layers using add()
model.add(Dense(512, activation='relu', input_shape=(x_train.shape[1] ,))) #1st layer, fully connected, 512 neurons (Dense) #number of features in input data
#relu- improves performance
model.add(Dropout(0.3)) #used to prevent overfitting by randomly dropping out some neurons during training
model.add(Dense(256, activation='relu')) #fully connected layer 2
model.add(Dropout(0.3))
model.add(Dense(128, activation='relu')) #fully connected layer 3
model.add(Dropout(0.3))
model.add(Dense(len(encoder.classes_), activation='softmax')) #dense layer, number of neurons= number of classes in the target variables
```

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(x_train, y_train, batch_size=64, epochs=100, validation_data=(x_test, y_test))
```

```python
train_accuracy = history.history['accuracy'][-1]
val_accuracy = history.history['val_accuracy'][-1]

print(f"Final training accuracy: {train_accuracy:.4f}")
```
